```python
x = data[['ram','px_height','battery_power','px_width','mobile_wt','int_memory','sc_w','talk_time','front_cam','sc_h']]
```

```python
x.head()
```

```python
x.shape
```

```python
x.describe()
```

```python
x = x.values
y = y.values
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
y = y.reshape(-1,1)
```

```python
# next converting y labels to one hot encoder
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder()
y = (oneHotEncoder.fit_transform(y)).toarray()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=101)
```

```python
#ANN model
import keras
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
model.add(Dense(8,activation='relu',input_dim = 10))
model.add(Dense(6,activation='relu'))
model.add(Dense(4,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=105,validation_data=(x_test,y_test),batch_size=64)
```

```python
y_pred = model.predict(x_test)
# inverse one hot encoding for x_test
pred = []
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))

# inverse encoding for y_test labels

test = []
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred,test)
print("Accuracy  : " , str(acc*100))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Features','Price_Weight'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Features','Price_Weight'],loc='upper left')
plt.show()
```
